```python
df_movies = pd.read_csv("..\Downloads\ml-25m\movies.csv")
df_movies.head()
```

```python
df_movies.drop_duplicates(['title'], keep='first', inplace=True)
df_movies.shape
```

```python
df_movies['title'] = df_movies['title'].str[:-7]
df_movies.head()
```

```python
df_movies['title'] = df_movies['title'].apply(lambda x: x.lower())
df_movies.head()
```

```python
df_ratings = pd.read_csv("/kaggle/input/movielens-20m-dataset/rating.csv")
df_ratings.head()
```

```python
df_ratings.isna().sum()
```

```python
df_ratings.shape
```

```python
# only movieId and title taken
df_movies = df_movies.loc[:, ['movieId','title']]
df_movies.head()
```

```python
df_ratings = df_ratings.loc[:,['userId','movieId','rating']]
df_ratings.head()
```

```python
df = pd.merge(df_movies, df_ratings)
df.head()
```

```python
pivot_table = df.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table.head()
```

```python
movie_watched = pivot_table["vampire in brooklyn"]
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
similarity_with_other_movies.head(10)
```
