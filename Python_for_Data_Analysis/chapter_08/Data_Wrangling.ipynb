### Combining and merging datasets

```python
df1 = pd.DataFrame({
    'key1':['a','b','c','b','a','a','c'],
    'data1':range(7),
})
df1
```

```python
df2 = pd.DataFrame({
    'key1':['a','b','d'],
    'data2':range(3),
})
df2
```

### merging df1 and df2

```python
pd.merge(df1,df2)
```

### specify which columns to merge

```python
pd.merge(df1,df2,on='key1')
```

### if column names are different

```python
df3 = pd.DataFrame({
    'key1':['a','b','c','b','a','a','c'],
    'data1':range(7),
})
df3
```

```python
df4 = pd.DataFrame({
    'key2':['a','b','d'],
    'data2':range(3),
})
df4
```

```python
pd.merge(df3,df4, left_on='key1', right_on='key2')
```

### merge takes intersection, union is available

```python
pd.merge(df1,df2, how='outer')
```

```python
df1 = pd.DataFrame({
    'key':['a','b','a','b','c','a','c'],
    'data1':range(7)
})
df1
```

```python
df2 = pd.DataFrame({
    'key':['a','b','a','a','d'],
    'data2':range(5),
})
df2
```

### left includes all the left elements

```python
pd.merge(df1,df2,on='key', how='left')
```

### right takes all the right elements

```python
pd.merge(df1,df2, on='key', how='right')
```

```python
pd.merge(df1,df2,on='key')
```

### to merge multiple columns

```python
df1 = pd.DataFrame({
    'key1':['a','b','c'],
    'data1':range(3),
    'key2':['foo','bar','foo']
})
df1
```

```python
pd.merge(df1,df2, on=['key1','key2'], how='outer')
```

### merging on Index

```python
left = pd.DataFrame({
    'key':['a','b','a','b','c'],
    'value':range(5),
})
left
```

```python
right=pd.DataFrame({
    'value':[3,5]
},
    index=['a','b']
)
right
```

```python
pd.merge(left,right, left_on='key', right_index=True)
```

```python
pd.merge(left,right, left_on='key', right_index=True, how='outer')
```

### melt method - inverse of pivot

```python
df = pd.DataFrame({
    'key':['foo','bar','baz'],
    'A':[0,1,2],
    'B':[1,2,3],
    'C':[2,3,4],
})
df
```

```python
res = pd.melt(df, ['key'])
res
```

```python
reshaped = res.pivot('key','variable','value')
reshaped
```

```python
reshaped.reset_index()
```
