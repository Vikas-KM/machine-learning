```python
x = np.array([1,2,3,4,5,6,7])
x
fig, ax = plt.subplots()
ax.plot(x, x**2, marker='o', label='squared line')
ax.legend()
```

### hspace and wspace - spacing between the subplots

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, x**2, color='g', linestyle='--')
ax[0,1].bar(x,np.exp(x), color='m')
ax[1,0].hist(x,x/2, color='r')
ax[1,1].scatter(x,x**3, color='y')
# plt.subplots_adjust(wspace=0, hspace=0) # comment this line and see
```

### xticks xtickslabels and xlim

```python
np.random.seed(1)
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum()); # remove semicolon and check the result
```

```python
np.random.seed(1)
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum()) 
ticks = ax.set_xticks([0,200,400,600,800,1000])
labels = ax.set_xticklabels(['one','two','three','four','five','six'], rotation=30)
```

```python
np.random.seed(1)
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0,200,400,600,800,1000])
labels = ax.set_xticklabels(['one','two','three','four','five','six'], rotation=30)
props={
    'title':'My First Plot',
    'xlabel':'Stages'
}
ax.set(**props)
```

```python
np.random.seed(1)
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum(), 'r--', label='one') 
ax.plot(np.random.randn(1000).cumsum(), 'g-.', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'b*', label='three')
ax.legend()
```

### Annotate

```python
np.random.seed(1)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(np.random.randn(1000).cumsum(), 'r--', label='one')
annotate_data = [(80,'one'), (260,'two'), (450,'three'), (760,'four')]
ax.annotate(s='one', xy=(80,10), xytext=(80,20), 
        arrowprops=dict(facecolor='black', headwidth=4, width=2,
            headlength=4),horizontalalignment='left', verticalalignment='top')
plt.savefig('annotate_plot.png', dpi=500)
```
