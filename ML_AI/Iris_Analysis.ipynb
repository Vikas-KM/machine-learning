```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

```python
sns.countplot(x='species', data=iris)
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
g = sns.FacetGrid(iris, hue='species', size=6)
g.map(plt.scatter, 'sepal_length', 'sepal_width')
g.add_legend()
```

```python
g = sns.FacetGrid(iris, hue='species', size=6)
g.map(plt.scatter, 'petal_length', 'petal_width')
g.add_legend()
```

```python
sns.pairplot(data=iris, hue='species' )
```

```python
sns.stripplot(x='species', y='petal_length', data=iris, hue='species', jitter=False)
```

```python
sns.stripplot(x='species', y='sepal_length', data=iris, hue='species', jitter=False)
```

```python
g = sns.FacetGrid(iris, hue='species')
g.map(sns.distplot, 'petal_length')
g.add_legend()
```
