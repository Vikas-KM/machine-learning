```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([2,4,6,3], label='myLine', ls='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('My Line')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y, color='red', marker='o')
plt.grid()
plt.show()
```

```python
x = np.arange(5)
y = x**2
z = x**3
plt.plot(x,y, label='^2', linewidth=5)
plt.plot(x,z, label='^3', linewidth=5)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,y,x,z)
plt.show()
```

```python
plots = plt.plot(x,y,x,z)
plt.setp(plots[0], color='red')
plt.setp(plots[1],color='yellow')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.grid()
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)


ax1.bar([1,2,3],[4,5,6])
ax2.barh([1,2,3],[4,5,6])

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)


ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter([3,2,1,5,4],[2,1,5,4,3])

plt.show()
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
# makes the plots fit together nicely
plt.tight_layout()
plt.show()
```

```python
## saving the plot as a figure
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
plt.savefig('line_graph.jpg')
```

```python
## saving to pdf file
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('multipage.pdf')

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()

pp.savefig()
pp.close()
```

### few functions to remember
* plt.cla() -> to clear the axis
* plt.clf() -> to clear the entire figure
* plt.close() -> to close the window

```python
x =  np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

### Functional method of matplotlib

```python
### Functional Method
plt.plot(x,y, label='Square', color='red')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.title('Functional method')
plt.show()
```

### OO method

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('X Values')
ax.set_ylabel('Y Values')
ax.set_title('Squared Line')
ax.plot(x,y, label='line')
ax.legend()
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y, 'r--')
ax2.plot(y,x, 'b^')

ax2.set_title('Smaller')
ax1.set_title('Bigger')
```

### creating subplots

### Saving the plots with dpi

### legend

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='square')
ax.plot(x, x**3, label='cube')
ax.plot(x, x**4, label='quad')
ax.legend()
plt.show()
```

### customizing the plot looks

```python
fig,ax = plt.subplots(1,1)
ax.plot(x,y, color='red', linestyle='-.', linewidth=5, alpha=0.5)
```

```python
fig,ax = plt.subplots(1,1)
ax.plot(x,y, color='red', ls='dashed', lw=0.5, alpha=0.8, marker='o', markersize=15,
       markerfacecolor='green', markeredgewidth=3, markeredgecolor='yellow')
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x,y)
#setting the xlimit and ylimit
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
