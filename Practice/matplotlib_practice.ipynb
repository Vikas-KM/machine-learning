```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([2,4,6,3], label='myLine', ls='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('My Line')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y, color='red', marker='o')
plt.grid()
plt.show()
```

```python
x = np.arange(5)
y = x**2
z = x**3
plt.plot(x,y, label='^2', linewidth=5)
plt.plot(x,z, label='^3', linewidth=5)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,y,x,z)
plt.show()
```

```python
plots = plt.plot(x,y,x,z)
plt.setp(plots[0], color='red')
plt.setp(plots[1],color='yellow')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.grid()
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
ax.set_ylim(8, 35)
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)


ax1.bar([1,2,3],[4,5,6])
ax2.barh([1,2,3],[4,5,6])

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)


ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter([3,2,1,5,4],[2,1,5,4,3])

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
# makes the plots fit together nicely
plt.tight_layout()
plt.show()
```

```python
## saving the plot as a figure
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
plt.savefig('line_graph.jpg')
```

```python
## saving to pdf file
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('multipage.pdf')

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot([1,2,3],[4,5,6], label='line')
ax.set_title('line graph')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()

pp.savefig()
pp.close()
```

### few functions to remember
* plt.cla() -> to clear the axis
* plt.clf() -> to clear the entire figure
* plt.close() -> to close the window
