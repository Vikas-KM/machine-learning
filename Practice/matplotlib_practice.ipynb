```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([2,4,6,3], label='myLine', ls='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('My Line')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y, color='red', marker='o')
plt.grid()
plt.show()
```

```python
x = np.arange(5)
y = x**2
z = x**3
plt.plot(x,y, label='^2', linewidth=5)
plt.plot(x,z, label='^3', linewidth=5)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,y,x,z)
plt.show()
```

```python
plots = plt.plot(x,y,x,z)
plt.setp(plots[0], color='red')
plt.setp(plots[1],color='yellow')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.grid()
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```
