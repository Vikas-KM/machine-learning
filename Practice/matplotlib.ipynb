### Matplotlib flow example

```python
# data prep
x = [1,2,3,4]
y = [11,22,33,44]

# setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width, height

# plot data
ax.plot(x,y)

# customize the plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# save and show
fig.savefig('simple_plot.png')
```

### two options for subplots

### subplots option 1

```python
# subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots( nrows=2,
                                            ncols=2,
                                        figsize=(10,10))
ax1.plot(x,x/2, color='r')
ax2.scatter(x, x**2, color='g')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color='y')
ax4.hist(np.random.randn(1000), color='m')
```

### subplots option 2

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,10))
ax[0,0].plot(x, x/1, color='r')
ax[0,1].scatter(x, np.exp(x), color='g')
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values(), color='y')
ax[1,1].hist(np.random.randn(1000), color='m')
```

### using OOP API of Matplotlib

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

### pyplot method

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

### OO Method mixed with pyplot

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
ax.set_xlim([45,80])
```

### OO method from scratch

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
#customize the plot
ax.set_title('Heart Disease and Cholesterol Levels')
ax.set_xlabel('age')
ax.set_ylabel('Cholesterol')
# ax.legend()
ax.legend(*scatter.legend_elements(), title='target')
# add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--', c='r')
```

```python
over_50.head()
```

```python
### subplots chol, age and thalach
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1,
                            figsize=(10,10), sharex=True)

# adding data to axes
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['chol'],
                     c=over_50['target'])
ax1.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['chol'].mean(), linestyle='--', c='r')

# adding data to ax2
hist = ax2.scatter(x=over_50['age'],
                y=over_50['thalach'],
               c=over_50['target'])
ax2.set(title='Heart Disease and Heart Rate Levels',
        xlabel='Age', ylabel='Heart Rate')
ax2.legend(*scatter.legend_elements(), title='Target')


# adding title to figure
fig.suptitle('heart Disease Analysis', fontsize=20)
```
