## Matplotlib tutorial

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
## Functional Method vs OOP Method
```

```python
# Functional Method
x = np.array([1,2,3,4])
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y Axis')
plt.title('Square Graph')
plt.show()
```

### Multiplot on the same canvas

```python
plt.subplot(1,2,1) # rows, columns and plot number
plt.plot(x,y, 'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'g--')

plt.show()
```

## OOP Method plotting

```python
# create figure object and call methods on that

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Square Graph')
axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

# the values should be within 0 and 1, check what happens if the values are more than 1?
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.6])
```

```python
fig = plt.figure()

# the values should be within 0 and 1, check what happens if the values are more than 1?
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.6])

axes1.plot(x,y)
axes2.plot(y,x)

plt.show()
```

```python
fig,axes = plt.subplots(2,2)
```

```python
x = np.array([1,2,3,4,5])
y = x**2
z = x**3
w = x**4
```

```python
fig,axes = plt.subplots(2,2)

axes[0][0].plot(x,y, 'r')

axes[1][0].plot(x,z, 'g')

axes[0][1].plot(x,w, 'b')

axes[1][1].plot(x,x, 'y')

plt.show()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
# change the figsize 
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='square')
ax.legend()
plt.show()
```
