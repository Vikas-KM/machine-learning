```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### distribution plot
* univariate means single value

kde = kernel density estimation

```python
sns.distplot(tips['total_bill'])
```

```python
# kde= false -> removes the curve
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], bins=30)
```

```python
sns.distplot(tips['total_bill'], bins='auto')
```

### Joint Plots
* two dist plots are combined

```python
x = tips['total_bill']
y = tips['tip']
sns.jointplot(x,y,kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y ='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

### GRIDS - customization

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.dtypes
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

### pairgrid

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='day')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='sex')
g.map(plt.scatter, 'total_bill', 'tip')
```
