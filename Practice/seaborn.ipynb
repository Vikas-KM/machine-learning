```python
import numpy as np 
import pandas as pd 
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
import seaborn as sns 
```

## Distribution Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
tips.size
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=10)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

## Joint Plots

```python
## combine two distplots
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

## Pair Plot

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

## rugplot

```python
sns.rugplot(tips['total_bill'])
```

## Kde Plot

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
sns.barplot(x='tip',y='smoker',data=tips)
```

```python
sns.barplot(x='tip', y='sex', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.countplot(y='sex', data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='smoker', split=True)
```

```python
sns.stripplot(x='smoker',y='tip',data=tips)
```

```python
sns.stripplot(x='smoker',y='tip',data=tips, jitter=False)
```

```python
sns.stripplot(x='smoker',y='tip',data=tips, hue='sex')
```

```python
sns.swarmplot(x='sex',y='tip',data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips)
sns.swarmplot(x='day',y='tip',data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

## Matrix Plots

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.shape
```

```python
# correlation
tips.corr()
```

```python
flights.corr()
```

```python
# data needs to be in matrix form for heatmap
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), cmap='YlGnBu')
```

```python
fp = flights.pivot_table(values='passengers', index='month', columns='year')
fp
```

```python
# using this data you can tell which year and month had more passengers
sns.heatmap(fp)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
