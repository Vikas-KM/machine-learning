# Plotting Techniques


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Line Chart
- Line plot is used to plot two or more continuous variables

```
plt.plot([1,2,3,4])
plt.ylabel('Some random numbers')
plt.show()
```

### Note
- You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].

```
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Squares of the number')
plt.show()
```

### Bewlo Two plots are same except the values on the x-axis and y-axis
- see the plt.axis ranges

### The axis function in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.

```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r--')
plt.axis([1, 4, 1, 16])
plt.xlabel('Values start from 1 to 4')
plt.ylabel('Values start from 1 to 16')
plt.title('How changing axis changes the scale of graph')
plt.show()
```

```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r--')
plt.axis([0, 6, 0, 20])
plt.xlabel('Values start from 0 to 6')
plt.ylabel('Values start from 0 to 20')
plt.title('How changing axis changes the scale of graph')
plt.show()
```

```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r--')
plt.axis([0, 10, 0, 30])
plt.xlabel('Values start from 0 to 10')
plt.ylabel('Values start from 0 to 30')
plt.title('How changing axis changes the scale of graph')
plt.show()
```

### Combining different plots into one

```
import numpy as np

t = np.arange(0, 5, 1)

# red dashes, blue squares dashes and green triangles dashes
plt.plot(t, t, 'r--', t, t**2, 'bs-', t, t**3, 'g^-')
plt.legend(['linear','square','cube'])
plt.show()
```

### working with ticks

```
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.xticks([0.5,1.5,2.5,3.5,4.5])
plt.title('Squares of the number')
plt.show()
```

### labelling of xticks can ne strings as well

```
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.xticks([0,1,2,3,4],['a','b','c','d','e'])
plt.title('Squares of the number')
plt.show()
```
